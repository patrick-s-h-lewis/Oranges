#Normalising Scrips
perform format changing operations on data in database to make it compatible with others

```python
import pymongo
from pymongo import MongoClient
mongo_url = 'mongodb://localhost:27017/'
db = 'BathSim'
coll_in = 'BathHarvest'
coll_out = 'BathAuthors'
client = MongoClient(mongo_url)
ci = client[db][coll_in]
co = client[db][coll_out]
```

```python
curs = ci.find()
```

```python
for rec in curs[0:1000]:
    print(rec['year'])
```

```python
punct_filter = dict((ord(char), u'') for char in ' "#$%&\'()*+,./-:;<=>?@[\\]^_`{|}')   
typical = [u'Abid, M. S.',
 u'Ahmed, S.',
 u'Ahmed, E.',
 u'Hussain, S.',
 u'Kiermaier, J.',
 u'Matysik, F.-M.',
 u'Mitchels, J.M.']
fin=[]
for i in typical:
    l,f =i.split(',')
    f = f.translate(punct_filter)
    fin.append(f+' '+ l)
```
